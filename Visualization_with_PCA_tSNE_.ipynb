### **Visualizing High-Dimensional Data**

Visualizing high-dimensional data is essential for exploratory data analysis and
understanding the underlying structure of the data. Feature extraction techniques such
as PCA and t-SNE provide effective ways to reduce dimensionality while preserving
important information.

1. Principal Component Analysis (PCA)

PCA is a linear dimensionality reduction technique that finds the principal components
(orthogonal directions) that maximize the variance in the data. It projects
high-dimensional data onto a lower-dimensional subspace while preserving as much
variance as possible.

2. t-Distributed Stochastic Neighbor Embedding (t-SNE)

t-SNE is a nonlinear dimensionality reduction technique that preserves local structure in
the data. It reduces dimensionality by modeling pairwise similarities between data points
in the high-dimensional space and mapping them to a lower-dimensional space.

#### **Example 1: Visualizing Iris Dataset with PCA**
Let us start by demonstrating with the popular iris data

```python
#import libraries

import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA
```

```python
# Load the Iris dataset
iris = load_iris()
X = iris.data
y = iris.target
```

```python
X.shape
```

```python
y.shape
```

```python
iris.target_names  #iris categorie's names
```

```python
# Perform PCA for dimensionality reduction
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)
```

```python
X_pca[:5]
```

```python
#Plot the reduced data
plt.figure(figsize=(8, 6))
for i in range(len(iris.target_names)):
    plt.scatter(X_pca[y == i, 0], X_pca[y == i, 1], label=iris.target_names[i])
    plt.title('PCA Visualization of Iris Dataset')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend()
plt.show()
```

#### **Example 2: Visualizing Iris Dataset with t-SNE**

```python
#Perform t-SNE for dimensionality reduction
from sklearn.manifold import TSNE

tsne = TSNE(n_components=2)
X_tsne = tsne.fit_transform(X)
```

```python
# Plot the reduced data
plt.figure(figsize=(8, 6))
for i in range(len(iris.target_names)):
    plt.scatter(X_tsne[y == i, 0], X_tsne[y == i, 1], label=iris.target_names[i])
plt.title('t-SNE Visualization of Iris Dataset')
plt.xlabel('t-SNE Component 1')
plt.ylabel('t-SNE Component 2')
plt.legend()
plt.show()
```

We can see that the vizualisation with PCA and tSNE in 2D is not the same.

###  **Example 3**

In this example, we apply the techniques of visualizing high-dimensional data and feature extraction to
another real dataset, the famous Breast Cancer Wisconsin (Diagnostic) dataset, which contains features computed from a digitized image of a fine needle aspirate (FNA) of a breast mass.
The features describe characteristics of the cell nuclei present in the image.

The goal is to use PCA and t-SNE to visualize this dataset in a lower-dimensional space.

```python
#Import Libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
```

```python
# Load the Breast Cancer dataset
cancer = load_breast_cancer()
X = cancer.data
y = cancer.target
```

```python
X[:3]
```

```python
y
```

```python
cancer.target_names
```

```python
#shape
print('X: ', X.shape)
print('y: ', y.shape)
```

```python
#Perform PCA for dimensionality reduction
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

# Perform t-SNE for dimensionality reduction
tsne = TSNE(n_components=2)
X_tsne = tsne.fit_transform(X)
```

```python
# Plot the reduced data using PCA
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
for i in range(len(cancer.target_names)):
    plt.scatter(X_pca[y==i, 0], X_pca[y==i, 1], label =  cancer.target_names[i], cmap=plt.cm.coolwarm, edgecolor='k')

plt.title('PCA Visualization of Breast Cancer Dataset')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend()
plt.show()
```

```python
# Plot the reduced data using t-SNE
plt.subplot(1, 2, 2)
for i in range(len(cancer.target_names)):
    plt.scatter(X_tsne[y==i, 0], X_tsne[y==i, 1], label = cancer.target_names[i], cmap=plt.cm.coolwarm, edgecolor='k')
plt.title('t-SNE Visualization of Breast Cancer Dataset')
plt.xlabel('t-SNE Component 1')
plt.ylabel('t-SNE Component 2')
plt.legend()
plt.tight_layout()
plt.show()
```

In summary:

- We load the Iris and Breast Cancer Wisconsin datasets.
- We perform PCA and t-SNE for dimensionality reduction, reducing the data to two dimensions.
- We plot the reduced Iris data using both PCA and t-SNE, with colors representing the target labels ( 'setosa', 'versicolor'and 'virginica).We did the same plot for the breast cancer data using both PCA and t-SNE, with colors representing the target labels (benign or malignant).
