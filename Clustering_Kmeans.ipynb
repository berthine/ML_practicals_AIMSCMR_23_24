### **Kmeans Clustering**

K-means clustering aims to partition $n$ observations into $k$ clusters in which each
observation belongs to the cluster with the nearest mean (centroid), serving as a
prototype of the cluster.

**Steps in K-means Clustering:**

1. Initialization: Select k initial centroids (the mean of the clusters).
2. Assignment step: Assign each data point to the nearest centroid.
3. Update step: Recalculate the centroids as the mean of all points in the cluster.
4. Repeat: Repeat steps 2 and 3 until the centroids no longer change significantly, meaning the model is converged.

```python
#import library
from sklearn.cluster import KMeans
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
```

For this first exemple, we are going to use the iris data from sklearn.

**a. Data loading**

```python
# Load the Iris dataset
iris = load_iris()
X = iris.data
y = iris.target
```

```python
#checking the lenght of the data
len(X)
```

**b. Model training**

```python
# Apply K-Means to cluster the data in 3 groups

kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X)
y_kmeans = kmeans.predict(X)
```

```python
#Print the K-means predictions
y_kmeans
```

```python
len(y_kmeans)
```

```python
##print the centroids

kmeans.cluster_centers_
```

**c. Visualisation**

```python
# Plotting the clusters
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans)
```

```python
iris.target_names
```

```python
centers = kmeans.cluster_centers_

plt.scatter(centers[:, 0], centers[:, 1], c='red', s=100, alpha=0.5, marker='X')
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans)
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')
plt.title('K-Means Clustering on the Iris Dataset')
plt.show()
```

The red cross indicate the centroid for each cluster.Each represents the mean location of data points in its cluster.

```python
#Elbow method to select the best k

sse = [ ]
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42).fit(X)
    sse.append(kmeans.inertia_)
plt.plot(range(1, 11), sse, marker='o')
plt.title('Elbow Method for Optimal K')
plt.xlabel('Number of clusters')
plt.ylabel('SSE')
plt.show()
```

From the elbow method, we can see that the optimal k is 3.

**d. Evaluation**

* Internal Evaluation Metrics:
Evaluate the algorithm based on the data itself without the need for
external information or ground truth labels. They measure the goodness of a clustering structure without reference to external knowledge. This include:

1. Inertia (Within-cluster Sum of Squares):
It measures the compactness of clusters, which is the sum of squared
distances of samples to their nearest cluster center. Lower inertia values indicate
better clusters that are dense and well-separated from other clusters. However, inertia is sensitive to the scale of the data, and lower values can sometimes be achieved by increasing the number of clusters, which may not always imply better clustering.


2. Silhouette Score:
It measures how similar an object is to its cluster (cohesion) compared to other clusters (separation). We want high cohesion and high separation. The silhouette coefficient combines ideas of both cohesion and
separation. The score ranges from -1 to 1, where a high value indicates that the object is well-matched to its cluster and poorly matched to neighboring clusters. If most objects have a high value, the clustering configuration is appropriate. If many points have a low or negative value, the clustering configuration may have too many or too few clusters.
It is computed as:

$$SC= (b-a)/max(a,b)$$


```python
#Let compute the inertia

inertia = kmeans.inertia_
print(inertia)
```

```python
from sklearn.metrics import silhouette_score
silhouette = silhouette_score(X, y_kmeans)
print(silhouette)
```

From the internal evaluation metrics, we can see that the inertia is high,  which means that the clusters are not well separeted. And for silhoute score, it is 0.55 which is a bit far to 1. This means, the points in on different clusters are not well-matched to its cluster and it match to neighboring clusters. In other words, the score of 0.55 indicate that, on average, each data point is closer to other points in its cluster than to points in other clusters.

Let's use external metrics such as ari, homogeneity and completeness to evaluate the predictions.

1. Adjusted Rand Index (ARI):
- An ARI of 1.0 indicates that the clustering perfectly matches the true labels.
This is an ideal scenario where the clusters formed by the algorithm exactly
correspond to the actual clusters in the data.

2. Homogeneity:
- A score of 1.0 means that all clusters contain only members of a single class.
This implies that each cluster is perfectly homogeneous concerning the true labels.

3. Completeness:
Similarly, a completeness score of 1.0 indicates that all members of a given
class are assigned to the same cluster. This means no class has been split across
multiple clusters.

```python
#import the library
from sklearn.metrics import silhouette_score, adjusted_rand_score,homogeneity_score, completeness_score
```

```python
silhouette = silhouette_score(data, predicted_labels)
ari = adjusted_rand_score(true_labels, predicted_labels)
homogeneity = homogeneity_score(true_labels, predicted_labels)
completeness = completeness_score(true_labels, predicted_labels)

#print results
print ('silhouette: ', silhouette,', ari:', ari,', homogeneity:', homogeneity, ', completeness:',completeness)
```

From the obtained values, we can see that the clustering perfectly matches the true labels (because of the ARI=1),
all clusters contain only members of a single class (because homogeneity=1) and all members of a given class are assigned to the same cluster (compleness = 1).

### Preprocessing
For simplicity, we focus on two features: 'Annual Income (k$)' and 'Spending

```python
X[:5]
```

#### Choosing the Number of Clusters

Before we can run K-means clustering, we need to decide on the number of clusters. One method to help determine this is the "elbow method", which involves plotting the within-cluster sum of squares (WCSS) against the number of clusters and looking for a 'knee' in the graph.

```python
wcss = [ ]
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

Based on the elbow method, we choose a number where the WCSS begins to decrease more slowly. In our case, the number is equal to 5.

#### Applying K-means Clustering

We use 5 clusters as suggested by the elbow method.

```python

kmeans = KMeans(n_clusters=5, init= 'k-means++', random_state=42)
y_kmeans = kmeans.fit_predict(X)
```

```python
y_kmeans
```

```python
# Visualize the clusters
#####################################################################
plt.figure(figsize=(10, 6))
sns.scatterplot(X[y_kmeans == 0, 0], X[y_kmeans == 0, 1], color = 'yellow', label =
'Cluster 1', s=50)
sns.scatterplot(X[y_kmeans == 1, 0], X[y_kmeans == 1, 1], color = 'blue', label =
'Cluster 2', s=50)
sns.scatterplot(X[y_kmeans == 2, 0], X[y_kmeans == 2, 1], color = 'green', label =
'Cluster 3', s=50)
sns.scatterplot(X[y_kmeans == 3, 0], X[y_kmeans == 3, 1], color = 'cyan', label =
'Cluster 4', s=50)
sns.scatterplot(X[y_kmeans == 4, 0], X[y_kmeans == 4, 1], color = 'magenta', label
= 'Cluster 5', s=50)
```

```python
# Plot the centroids of the clusters
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=100,
c='red', label='Centroids')
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis')

plt.title('Clusters of customers')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend()
plt.show()
```
